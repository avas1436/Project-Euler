solving with Inefficient algorithm

```python
count = 0
for one in range(201): # 1p
    p1 = (one * 1) 
    if p1 > 200: break
    for tow in range(101): # 2p
        p2 = (tow * 2) + p1
        if p2 > 200: break
        for five in range(41): # 5p
            p5 = (five * 5) + p2
            if p5 > 200: break
            for ten in range(21): # 10p
                p10 = (ten * 10) + p5
                if p10 > 200: break
                for tewnty in range(11): # 20p
                    p20 = (tewnty * 20) + p10
                    if p20 > 200: break
                    for fifty in range(5): # 50p
                        p50 = (fifty * 50) + p20
                        if p50 > 200: break
                        for one_hundred in range(3): # 100p
                            p100 = (one_hundred * 100) +  p50
                            if p100 > 200: break
                            for tow_hundred in range(2): # 200p
                                p200 = (tow_hundred * 200) + p100
                                if p200 > 200: break
                                if p200 == 200:
                                    count += 1

print(count)
```
