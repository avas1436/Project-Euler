```python
from functools import cache
```

```python
@cache
def is_prime(num: int):
    if num < 2:
        return False
    if num == 2:
        return True
    if num % 2 == 0:
        return False
    i = 3
    while i * i <= num:
        if num % i == 0:
            return False
        i += 2
    return True
```

```python
def rotator(num: int) -> set[int]:
    s = str(num)
    return {int(s[j:] + s[:j]) for j in range(len(s))}
```

```python
def main(start: int= 1, end:int = 100):
    cache = set()
    for i in range(start, end+1):
        if i in cache:
            continue
        if not is_prime(i):
            continue
        nums = rotator(num=i)
        flag = True
        for j in nums:
            if not is_prime(j):
                flag = False
                break
        if flag:
            cache.update(nums)
            

    return cache
```

```python
answare = main(end=1_000_000)
print(len(answare))
```
