```python
def is_prime(num: int):
    if num < 2:
        return False
    if num == 2:
        return True
    if num % 2 == 0:
        return False
    i = 3
    while i * i <= num:
        if num % i == 0:
            return False
        i += 2
    return True
```

```python
def rotator(num: int):
    rotate_set = set()
    str_num = str(num)
    i = len(str_num) 
    for j in range(i):
        rotate_set.add(int(str_num[j:] + str_num[:j]))
    
    return rotate_set
```

```python
def main(start: int= 1, end:int = 100):
    cache = set()
    for i in range(start, end+1):
        if i in cache:
            continue
        nums = rotator(num=i)
        if all(is_prime(j) for j in nums):
            cache.update(nums)

    return cache
```

```python
answare = main(end=1_000_000)
print(len(answare))
```
