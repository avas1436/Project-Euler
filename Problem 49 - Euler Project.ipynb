```python
from math import pow
from functools import cache
```

```python
@cache
def is_prime(num: int):
    if num < 1 :
        return False
    if num == 2 :
        return True
    if num % 2 == 0 :
        return False
    limit = int(pow(num, 0.5)) + 1
    for i in range(3, limit):
        if num % i == 0:
            return False
    return True
```

```python
def prime_tuple(start: int, end: int):
    ls = []
    for i in range(start, end+1):
        if is_prime(i):
            ls.append(i)
    return tuple(ls)
```

```python
@cache
def same_permutations(my_tuple: tuple):
    final_dic = {}
    limit = len(my_tuple)
    for i, num in enumerate(my_tuple):
        final_dic[i] = {num}
        str_num = str(num)
        tuple_num = tuple(sorted(tuple(char for char in str_num)))
        for j in range(i+1, limit):
            str_num_2 = str(my_tuple[j])
            tuple_num_2 = tuple(sorted(tuple(char for char in str_num_2)))
            if tuple_num == tuple_num_2 :
                final_dic[i].add(my_tuple[j])
        if len(final_dic[i]) < 3:
            final_dic.pop(i)

    return final_dic
```

```python
def solution(final_dic: dict):
    answare = set()
    for group in final_dic.values():
        ls = sorted(group)
        n = len(ls)
        
        # بررسی همه سه‌تایی‌های ممکن
        for i in range(n):
            for j in range(i + 1, n):
                for k in range(j + 1, n):
                    if ls[j] - ls[i] == ls[k] - ls[j]:
                        diff = ls[j] - ls[i]
                        if diff > 0:  # مطمئن شویم دنباله افزایشی است
                             ans = f"{ls[i]},{ls[j]},{ls[k]}"
                             answare.add(ans)
    return answare
```

```python
digit_tuple = prime_tuple(1000, 9999)
same_permutations_dic = same_permutations(digit_tuple)
solution(same_permutations_dic)
```
