```python
def find_palindrome(start: int, end: int):
    ls = []
    for num in range(start, end+1):
        if len(str(num)) > 1 :
            if int(str(num)[::-1]) == num:
                ls.append(num)
    return ls
```

```python
s = 100 * 100
e = 999 * 999
numbers_ls = find_palindrome(s, e)
```

```python
def answare(numbers: list):
    ls = []
    for num in numbers:
        if (999//num) - ((100 + num - 1) // num)  >= 1:
            ls.append(num)
    return ls
```

```python
answare(numbers_ls)
```
