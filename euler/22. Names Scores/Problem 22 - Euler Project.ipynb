```python
from functools import cache
```

```python
def word_number(word: str):
    value = 0
    for char in word:
        value += (ord(char.upper()) - ord("A") + 1)
    return value
```

```python
def file_content():
    with open(r"words.txt", "r") as file:
        content = file.read()
        content_tuple = tuple(sorted(content.replace('"', '').split(",")))
        return content_tuple
```

```python
content = file_content()
print(len(content))
```

```python
@cache
def answare(content: tuple):
    total = 0
    for i, word in enumerate(content, 1):
        word_score = word_number(word)
        total += (word_score * i)
    return total
```

```python
answare(content)
```
