```python
def answare(start: int, end: int):
    pal_list = []
    for num in range(start, end+1):
        if num > 10:
            if int(str(num)[::-1]) == num:
                pal_list.append(num)
    final_ls = []
    for number in pal_list:
        for i in range(999, 100, -1):
            if number % i == 0:
                if 100 < number // i < 1000:
                    final_ls.append(number)  
    return max(final_ls)        
    
```

```python
if __name__ == "__main__":
    s = 100 * 100
    e = 999 * 999
    print(answare(start=s, end=e))
```
