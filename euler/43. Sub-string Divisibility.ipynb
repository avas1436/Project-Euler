```python
def find_pandigits() -> set:
    pandigits = []
    # عدد اول میتواند همه اعداد به غیر از 0 باشد
    d1_tup = (1, 2, 3, 4, 5, 6, 7, 8, 9)

    # عدد چهارم میتواند یکی از اعداد لیست باشد
    d4_tup = (0, 2, 4, 6, 8)

    # عدد ششم باید 0 و 5 باشد تا کل عدد قابل قسمت تعیین شده بر 5 بخش پذیر باشد
    d6_tup = (0, 5)

    # بقیه اعداد باید بچرخند تا تعیین شود نتیجه چیست
    others = (1, 0, 2, 3, 4, 5, 6, 7, 8, 9)

    for d1 in d1_tup:
        for d2 in others:
            if d2 in [d1]:
                continue
            for d3 in others:
                if d3 in [d1, d2]:
                    continue
                for d4 in d4_tup:
                    if d4 in [d1, d2, d3]:
                        continue
                    for d5 in others:
                        if d5 in [d1, d2, d3, d4]:
                            continue
                        if (d3 + d4 + d5) % 3 != 0:
                            continue
                        for d6 in d6_tup:
                            if d6 in [d1, d2, d3, d4, d5]:
                                continue
                            for d7 in others:
                                if d7 in [d1, d2, d3, d4, d5, d6]:
                                    continue
                                if int(f"{d5}{d6}{d7}") % 7 != 0:
                                    continue
                                for d8 in others:
                                    if d8 in [d1, d2, d3, d4, d5, d6, d7]:
                                        continue
                                    if int(f"{d6}{d7}{d8}") % 11 != 0:
                                        continue
                                    for d9 in others:
                                        if d9 in [d1, d2, d3, d4, d5, d6, d7, d8]:
                                            continue
                                        if int(f"{d7}{d8}{d9}") % 13 != 0:
                                            continue
                                        for d10 in others:
                                            if d10 in [d1, d2, d3, d4, d5, d6, d7, d8, d9]:
                                                continue
                                            if int(f"{d8}{d9}{d10}") % 17 != 0:
                                                continue
                                            pandigits.append(
                                                int(f"{d1}{d2}{d3}{d4}{d5}{d6}{d7}{d8}{d9}{d10}")
                                            )
    
    return set(pandigits)
```

```python
numbers = find_pandigits()
print(sum(numbers))
```
