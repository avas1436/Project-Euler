```python
from math import pow
from functools import cache
```

```python
@cache
def power(num: int):
    total = 0
    for i in range(1, num+1):
        total += i ** i
    limit = len(str(total))
    if limit > 10:
        place = limit - 10
        total_str = str(total)[place:]    
    return total_str
```

```python
print(power(1000))
```
