```python
def letter_counter(num: int) -> int:
    letter_counts = {
        0:0,
        1: 3,
        2: 3,
        3: 5,
        4: 4,
        5: 4,
        6: 3,
        7: 5,
        8: 5,
        9: 4,
        10: 3,
        11: 6,
        12: 6,
        13: 8,
        14: 8,
        15: 7,
        16: 7,
        17: 9,
        18: 8,
        19: 8,
        20: 6,
        30: 6,
        40: 5,
        50: 5,
        60: 5,
        70: 7,
        80: 6,
        90: 6,
    }
    
    if 0 <= num <= 19:
        return letter_counts[num]
    
    elif 20 <= num <= 99:
        units = letter_counts[int(str(num)[1])]
        tens = letter_counts[int(str(num)[0]) * 10]
        return units + tens
    
    elif 100 <= num <= 999:
        units = letter_counts[(num // 100)]
        if num % 100 == 0:
            return units + 7
        specials = int(str(num)[1:])
        return units + 7 + 3 + letter_counter(specials)
    elif num == 1000:
        return 11
```

```python
count = 0
for i in range(1,1001):
    count += letter_counter(i)

print(count)
```
